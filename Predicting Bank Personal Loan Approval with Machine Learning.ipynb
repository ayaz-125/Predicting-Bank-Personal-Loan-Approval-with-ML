### SPLITTING THE DATASET INTO TARGET VARIABLE AND INDEPENDENT VARIBALE

```python
x = data.iloc[:,1:13]
```

```python
x
```

```python
y = data.iloc[:,13]
```

```python
y = pd.DataFrame(y)
```

```python
y
```

### SPLITING THE DATASET INTO TRAINING AND TESTING PART

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=1,test_size=0.20)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

### BUILDING THE MODEL WITH THE HELP OF LOGISTIC REGRESSION

```python
model1 = LogisticRegression()
model1.fit(x_train,y_train)
```

```python
y_predict = model1.predict(x_test)
```

```python
y_predict
```

```python
len(y_predict)
```

```python
print("Training score of the model--->",model1.score(x_train,y_train)*100)
print("Testing score of the model -->",model1.score(x_test,y_test)*100)
```

```python
from sklearn import metrics
cm = metrics.confusion_matrix(y_test,y_predict,labels=[1,0])
dm_cm = pd.DataFrame(cm,index = [i for i in ["1",'0']],
                                columns = [i for i in ['Predicted 1','Predicted 0']])
plt.figure(figsize = (7,5))
sns.heatmap(dm_cm,annot = True)
plt.show()
```

### BUILDING MODEL WITH THE HELP OF KNN ALGORITHM

```python
from sklearn.neighbors import KNeighborsClassifier
model2 = KNeighborsClassifier(n_neighbors=3)
model2.fit(x_train,y_train)
```

```python
y_predict_knn = model2.predict(x_test)
```

```python
y_predict_knn
```

```python
y_predict_knn.shape
```

```python
print("Training score of the model ---->",model2.score(x_train,y_train)*100)
print("Testing score of the model --->",model2.score(x_test,y_test)*100)
```

```python
cm2 = metrics.confusion_matrix(y_test,y_predict_knn,labels=[1,0])
dm_cm2 = pd.DataFrame(cm2,index = [i for i in ["1","0"]],
                     columns = [i for i in ['predicted 1','predicted 0']])
plt.figure(figsize = (7,5))
sns.heatmap(dm_cm2,annot = True)
plt.show()
```

### BUILDING THE MODEL WITH THE HELP OF NAIVE BAYES ALGORITHM

```python
from sklearn.naive_bayes import GaussianNB
model3 = GaussianNB()
model3.fit(x_train,y_train)
```

```python
y_predict_gb = model3.predict(x_test)
```

```python
y_predict_gb
```

```python
y_predict_gb.shape
```

```python
print("Training score of the model --->",model3.score(x_train,y_train)*100)
print("Testing score of the model --->",model3.score(x_test,y_test)*100)
```

```python
cm3 = metrics.confusion_matrix(y_test,y_predict_gb, labels=[1,0])
dm_cm3 = pd.DataFrame(cm3,index=[i for i in ["1","0"]],
                     columns = [i for i in ["Predicted 1","Predicted 0"]])
sns.heatmap(dm_cm3,annot=True)
plt.show()
```
